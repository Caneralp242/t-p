# MACHİNE LEARNİNG

```python
df=pd.read_csv("/kaggle/input/nba-1920-season-stats-ratings-and-salaries/nba-stats-salary-rating.csv")
df.drop(["Unnamed: 0","Player","Salaries"],axis=1,inplace=True)
#df=df.replace(np.nan,0)
#df=pd.get_dummies(df,columns=["Tm"],prefix=["Tm"])
#df=pd.get_dummies(df,columns=["Pos"],prefix=["Pos"])
```

```python
df.columns
```

```python
df.Ratings=[1 if each>df.Ratings.mean()  else 0 for each in df.Ratings]
y=df.Ratings.values
xt=df.drop(["Ratings"],axis=1)
from sklearn.preprocessing import StandardScaler
x=StandardScaler().fit_transform(xt)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
k=list(range(1,50))
w=["uniform","distance"]
grid=dict(n_neighbors=k,weights=w)
knn=KNeighborsClassifier()
cv=GridSearchCV(knn,grid,cv=10)
cv.fit(x_train,y_train)
print(cv.best_params_)
print(cv.best_score_)
```

```python
x=1
epoch=1
while True:
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(C=x,max_iter=150)
    log.fit(x_train,y_train)
    print("Train score: %",log.score(x_train,y_train)*100)
    print("Test score: %",log.score(x_test,y_test)*100)
    print("epochs:",epoch)
    if np.abs(log.score(x_train,y_train)-log.score(x_test,y_test))>0.1:
        print("model is overfitted")
        print("x:",x)
        x+=1
        epoch+=1
        
    else:
        break
        
    if log.score(x_test,y_test)<0.9:
        print("ov şit")
        print("x:",x)
        x+=1
        epoch+=1
  

    if log.score(x_train,y_train) <0.70 or log.score(x_test,y_test) <0.70:
        print("model is underfitted")
```
